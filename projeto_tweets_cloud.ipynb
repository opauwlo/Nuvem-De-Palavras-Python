```python
from wordcloud import WordCloud
from PIL import Image
import numpy as np
import os
DATA_DIR = "./output_clean_text"
OUT_PUT = "./output_img"
```

```python
def get_name(filename, dirname = DATA_DIR):
  return os.path.join(dirname, filename)

def out_put(filename, dirname = OUT_PUT):
  return os.path.join(dirname, filename)

def get_img(filename, dirname = OUT_PUT):
  return os.path.join(dirname,filename)
```

```python
tweetsfile = open(get_name("INEP.txt"))
text = tweetsfile.read()
tweetsfile.close()
```

```python
def simple(filename):
  wordcloud = WordCloud(width=2000, height=1000, background_color="white",
                        max_words=200,contour_color='steeblue',
                        min_word_length=3).generate(text)

  image = wordcloud.to_image()
  image.save(out_put(filename))
```

```python
def masked(filename, maskname):
  enem = np.array(Image.open(get_img("enem_mask.png")))
  wordcloud = WordCloud(width=2000, height=1000, background_color="white",
                        max_words=200,repeat=False,mask=enem,
                        contour_color='steelblue',contour_width=3).generate(text)

  image = wordcloud.to_image()
  image.save(out_put(filename))
```

```python
masked("enem_masked.png", "enem_mask.png")
simple("enem_simple.png")
```
