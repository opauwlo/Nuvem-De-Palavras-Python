```python
from wordcloud import WordCloud
import os
```

```python
tweetsfile = open(os.path.join("/content/drive/MyDrive/Colab Notebooks/output_clean_text", "liminha.txt"))
text = tweetsfile.read()
tweetsfile.close()
```

```python
wordcloud = WordCloud(width=2000, height=1000).generate(text)
```

```python
image = wordcloud.to_image()
image.save(os.path.join("/content/drive/MyDrive/Colab Notebooks/output_img", "liminha2.png"))
```
