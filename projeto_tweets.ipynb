```python
from searchtweets import ResultStream, gen_request_parameters, load_credentials
from searchtweets import collect_results
from wordcloud import WordCloud
import os
```

```python
stream_arg = load_credentials(filename="/content/drive/MyDrive/Colab Notebooks/config.yaml",
                 yaml_key="search_tweets_v2",
                 env_overwrite=False)
```

```python
OUT_DIR = "/content/drive/MyDrive/Colab Notebooks/output_text"
LIMIT_PER_CALL = 100
LIMIT_MAX = 1000
termo_de_busca = "Python"
```

**Aqui eu defino algumas variaveis.**


---




> OUT_DIR



**OUT_DIR** se refere ao diretoria de saida do arquivo que nosso script ira gerar.

---


> LIMIT



**LIMIT** se refere a quatidade de tweets por requisição, no meu e provavelmente no seu caso o twitter permite até **no máximo 100 tweets por requisição**.

---



> LIMIT_MAX


**LIMIT_MAX** se refere ao maximo de tweets que voce quer armazenar.

---



> termo_de_busca


**termo_de_busca** é a palavra chafe que você quer buscar.

```python
for day in range(01,08):
  query = gen_request_parameters(f"{termo_de_busca} -is:retweet lang:pt", 
                               start_time = f"2021-03-{day}",
                               results_per_call=LIMIT)
  
  tweets = collect_results(query,
                         max_tweets=LIMIT_MAX,
                         result_stream_args=stream_arg)
  
  print(f"{len(tweets)} resultados no dia {day}")
 
  with open(os.path.join(OUT_DIR, f"tweets_{termo_de_busca}.txt"), "a") as tweetsfile: 
    lines = [tweet['text'] for tweet in tweets if 'text' in tweet]
    tweetsfile.writelines(lines)
print("FIM")
```

**Função Principal que retorna o arquivo com os tweets**
