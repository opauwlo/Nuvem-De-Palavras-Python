```python
import os
```

```python
DATA_DIR = "/content/drive/MyDrive/Colab Notebooks/output_text"
OUTPUT_DIR = "/content/drive/MyDrive/Colab Notebooks/output_clean_text"
TO_REMOVE = ['(',')','[',']',',','.','?','!','*',':','/','\\'] 
```



---



> **TO_REMOVE**

Lista de caracteres que serão substituidos por espeços " ".

---






```python
def get_name(filename, dirname = DATA_DIR):
  return os.path.join(dirname, filename)

def stopwordslist():
  lines = []
  with open (get_name("stopwords.txt"), "r") as stopfile:
    lines = [ line.strip() for line in stopfile.readlines() ]
  return lines

def remove_chars(text, undesireds):
  for c in undesireds:
    text = text.replace(c," ")
  return text
```



---


> **def get_name**

Função simples, me retornar o nome do arquivo e diretoria. É só para eu e você não ter que ficar digitando muita coisa.

---



> **def stopwordslist**

Nos links de apoio está o repositorio do humano que criou está lista de stopwords(stopwords são palavras que não fazem sentido ou não prestam pra gente, conectivos e etc). A função quebra as linhas e remove os espaços e os /n devolvendo uma lista limpa das stopwords. *(experimente visualizar as stopwords sem esta formatação)*

---


> **def remove_chars**

Função simples que remove os caracteres que definimos no **TO_REMOVE**

---









```python
if __name__ == "__main__":
  with open(get_name("tweets_ENEM.txt")) as tweetsfile:
    content = tweetsfile.read()
    content = remove_chars(content, TO_REMOVE)
    parts = [p.strip() for p in content.split()]

    stopwords = stopwordslist()
    words = [part for part in parts if part not in stopwords]
    words = [w for w in words if not w.startswith("http")]
    words = [w for w in words if not w.startswith("@")]
    words = [w for w in words if not w.startswith("#")]

    with open(get_name("ENEM.txt", OUTPUT_DIR), "w") as textfile:
      textfile.write(" ".join(words))
    
```
